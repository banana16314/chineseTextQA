```python
dict = {}
ans = {}
wrong = {}
with open('BoP2017-DBQA.train.txt', 'r',encoding='utf-8') as f:
    for line in f:
        s = line.split('\t')
        dict[s[1]] = []
        wrong[s[1]] = []
        if s[0].isdigit() and int(s[0]) == 1:
            ans[s[1]] = s[2]
        elif s[0].isdigit() and int(s[0]) == 0:
            wrong[s[1]].append(s[2])
for q in dict:
    print(q, ans[q])
```

```python
WHAT = 0X01
WHERE = 0X02
WHEN = 0X03
WHY = 0X04
for q in dict:
    if (q.find('什么时候') == -1 and q.find('什么地方') == -1 and q.find('什么') != -1) or q.find('多少') != -1 or q.find('几') != -1 or q.find('谁') != -1 \
         or q.find('多') != -1 or q.find('是？') != -1 or (q.find('哪') != -1 and q.find('哪里') == -1 and q.find('哪儿') == -1 and q.find('哪一年') == -1) \
         or q.find('怎么') != -1 or q.find('是否') != -1 or q.find('如何') != -1 or q.find('吗？') != -1 or q.find('有没有') != -1 or q.find('怎样') != -1 \
         or q.find('有何') != -1 or q.find('是何') != -1 or q.find('为？') != -1:
        dict[q].append(WHAT)
    if q.find('哪里') != -1 or q.find('哪儿') != -1 or q.find('什么地方') != -1:
        dict[q].append(WHERE)
    if q.find('什么时候') != -1 or q.find('何时') != -1 or q.find('哪一年') != -1 or q.find('啥时候') != -1 or q.find('时间') != -1 or q.find('日期') != -1 or q.find('生日') != -1:
        dict[q].append(WHEN)
    if (q.find('为什么') != -1 and q.find('为什么？') == -1) or q.find('为何') != -1:
        dict[q].append(WHY)
```

```python
import thulac   

thu1 = thulac.thulac()  #默认模式

for q in dict:
    if WHEN in dict[q]:
        print('Q: ', q)
        text = thu1.cut(ans[q], text=True)
        print('A: ', ans[q])
        for i in text.split(' '):
            if i.find('_t') != -1:
                print(i)
        print('Wrong:')
        ctr = 0
        for w in wrong[q]:
            text = thu1.cut(w, text=True)
            for i in text.split(' '):
                if i.find('_t') != -1:
                    ctr += 1
                    if ctr != 0:
                        print('Ex: %s, word: %s' % (w, i))
            print('Ctr: ', ctr)
        print('\n')
        print('\n')
        print('\n')
```

```python
import thulac
import pickle
import sys

def getQuestionWordAndPoint(thu1, q):
    qw = thu1.cut(q, text=True).split(' ')
    pointQuestionTmp = list(filter(lambda x: '_n' in x or '_v' in x, qw))
    pointQuestion = []
    pointQuestionTmpLen = len(pointQuestionTmp)
    for i, x in enumerate(pointQuestionTmp):
        if '_n' in x: 
            if len(x) >= 4 and i >= 1 and i <= pointQuestionTmpLen - 2 \
               and list(filter(lambda x: '_v' in x, pointQuestionTmp[i - 1:i + 2])):
                pointQuestion.append(x)
        if '_v' in x and len(x) >= 4:
            pointQuestion.append(x)
    return (qw, pointQuestion)

def getQuestionType(q):
    res = []
#     if (q.find('什么时候') == -1 and q.find('什么地方') == -1 and q.find('什么') != -1) or q.find('多少') != -1 or q.find('几') != -1 or q.find('谁') != -1 \
#          or q.find('多') != -1 or q.find('是？') != -1 or (q.find('哪') != -1 and q.find('哪里') == -1 and q.find('哪儿') == -1 and q.find('哪一年') == -1) \
#          or q.find('怎么') != -1 or q.find('是否') != -1 or q.find('如何') != -1 or q.find('吗？') != -1 or q.find('有没有') != -1 or q.find('怎样') != -1 \
#          or q.find('有何') != -1 or q.find('是何') != -1 or q.find('为？') != -1:
#         res.append(WHAT)
    if q.find('谁') != -1 or q.find('何人') != -1 or q.find('什么人') != -1:
        res.append(WHO)
    if q.find('哪里') != -1 or q.find('哪儿') != -1 or q.find('什么地方') != -1:
        res.append(WHERE)
    if q.find('什么时候') != -1 or q.find('何时') != -1 or q.find('哪一年') != -1 or q.find('哪年') != -1 or q.find('啥时候') != -1 or q.find('时间') != -1 or q.find('日期') != -1 or q.find('生日') != -1:
        res.append(WHEN)
    if (q.find('为什么') != -1 and q.find('为什么？') == -1) or q.find('为何') != -1:
        res.append(WHY)
    if q.find('多') != -1 or q.find('几') != -1:
        res.append(AMOUNT)
    return res

def matchQuestionAndAnswer(questionType, q, answerWord):
    record = 0
    typeScore = 8
    wordScore = 8
    pointScore = 3
    wordWhere = None
    wordWhereTimes = 3
    wordWhen = None
    wordWho = None
    wordAmount = None
    questionWord = q[0]
    questionWordTmp = questionWord[:]
    
    if WHERE in questionType:
        wordWhere = '_ns'
    if WHEN in questionType:
        wordWhen = '_t'
    if WHO in questionType:
        wordWho = '_np'
    if AMOUNT in questionType:
        wordAmount = '_m'
    
    reversedNWordList = [x for x in questionWord if '_n' in x][::-1]
    if reversedNWordList and len(reversedNWordList) != 0:
        reversedNWord = reversedNWordList[0]
    else:
        reversedNWord = None
        
    reversedVWordList = [x for x in questionWord if '_v' in x and '有' not in x and '是' not in x][::-1]
    if reversedVWordList and len(reversedVWordList) != 0:
        reversedVWord = reversedVWordList[0]
    else:
        reversedVWord = None
   
    for word in answerWord:
        if wordWhere and wordWhere in word:
            record += typeScore
            wordWhereTimes -= 1
            if wordWhereTimes == 0:
                wordWhere = None
        if wordWhen and wordWhen in word:
            record += typeScore
            wordWhen = None
        if WHY in questionType and ('因为' in word or '因' in word):
            record += 10
        if wordAmount and wordAmount in word:
            record += 5
            wordAmount = None
        if wordWho and wordWho in word:
            record += 10
            wordWho = None
        if reversedNWord and reversedNWord == word:
            record += wordScore
            reversedNWord = None
        if reversedVWord and reversedVWord == word:
            record += wordScore
            reversedVWord = None
        if word in questionWordTmp:
            if word in q[1]:
                record += pointScore
            else:
                record += 1
            questionWordTmp = list(filter(lambda a: a != word, questionWordTmp))
#         if WHERE in questionType:
#             print(word, record)
    return record

WHO = 0X01
WHERE = 0X02
WHEN = 0X03
WHY = 0X04
AMOUNT = 0x05

#提取所有问题
fileName = 'BoP2017-DBQA.train.txt'
answerDict = {}

#关键词匹配
num = 0
matchNum = 0
stdAnswer = None
thu1 = thulac.thulac()  #默认模式

f = open(fileName,'r',encoding='utf-8')             # 返回一个文件对象
line = f.readline()             # 调用文件的 readline()方法
while line:
    lineList = line.split('\t')
    question = lineList[1]
    # 如果第一次遇到这个问题
    if question not in answerDict:
        answerDict[question] = [getQuestionWordAndPoint(thu1, question)]
    labelAnswerTuple = (lineList[0], lineList[2])
    answerDict[question].append(labelAnswerTuple)
    line = f.readline()
f.close()
with open('trainDict.pkl', 'wb') as output:
    pickle.dump(answerDict, output, pickle.HIGHEST_PROTOCOL)

# with open('trainDict.pkl', 'rb') as input:
#     answerDict = pickle.load(input)
#{q:[(qw,pointq),(0/1,answer)],}
for i in answerDict:
    if not i:
        continue
    maxRecord = 0
    stdRecord = 0
    bestAnswer = None
    
    if num < len(answerDict):
        num += 1
        #问题提取关键词
        questionType = getQuestionType(i)
#         print('T:%r\nQ:%s\n' % (questionType, i))
        #print "Q:"+i
        for j in (answerDict[i])[1:]:
            #回答提取关键词
            answerWord = list(set(thu1.cut(j[1], text=True).split(' ')))
            #获取最高匹配的答案
            record =  matchQuestionAndAnswer(questionType, answerDict[i][0], answerWord)
            if record > maxRecord:
                maxRecord = record
                bestAnswer = j[1]
            #返回正确答案
            if j[0] == '1':
                stdAnswer = j[1]
                stdRecord = record
                #print "StdA:"+str(record)+" " + j[1]
        if stdAnswer == bestAnswer:
            matchNum += 1
        else:
            print('Q:%s' % thu1.cut(i, text=True))
            print('QPW:%s\n%s' % answerDict[i][0])
            if not bestAnswer:
                continue
            print('W:%s' % thu1.cut(bestAnswer, text=True))
            print('WS:%d' % maxRecord)
            if not stdAnswer:
                continue
            print('A:%s' % thu1.cut(stdAnswer, text=True))
            print('AS:%d' % stdRecord)
            print('match:%d, total:%d, acc:%f' % (matchNum, num, matchNum / num))
            print('\n\n\n')
        #print "myA:"+str(maxRecord) +" "+ bestAnswer
        #print
print('match:%d, total:%d, acc:%f' % (matchNum, num, matchNum / num))
```
